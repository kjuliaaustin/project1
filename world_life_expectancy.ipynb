#Life Expectancy Related to Mental Health

```python
df2 = pd.read_csv('export_world_life_expectancy.csv')
df2.head()
```

```python
sns.set(font_scale=0.8)
features2=['Schizophrenia', 'Bipolar_disorder', 'Eating_disorder', 'Anxiety_disorder', 'Drug_use_disorder', 'Depression', 'Alcohol_use_disorder', 'Avg_Life_Expectancy']
sns.heatmap(df2[features2].corr(),
            annot = True,
            annot_kws={"size": 10},
            fmt = '.1g',
            center = 0,
            cmap = 'coolwarm',
            linewidths = 1,
            linecolor = 'black')
```

```python
y = df2['Avg_Life_Expectancy']
X = df2.Depression
```

```python
X = sm.add_constant(X)
X.head(2)
```

```python
model2 = sm.OLS(endog= y, exog = X).fit()
print(model2.summary())
```

A negative coefficient indicates a negative relationship. As depression increases, life expectancy decreases.

The p value is 0.12 indicating that there is not statistical significance between the relationship.

Since the R-squared is 0.015, 2% of the variance in life expectancy can be explained by depression.



*   Longer life expectancy often correlates with better overall health, which can lead to a higher quality of life and reduced physical pain and discomfort, all of which can contribute to lower rates of depression.

*   Older adults may develop greater cognitive resilience over time, enabling them to cope more effectively with life's challenges and reducing the risk of depression.

*   With age comes a broader perspective and greater wisdom, which can help individuals navigate difficult life events and reduce the impact of stressors on mental health.

*   Depression is associated with changes in the body's immune and inflammatory responses, which can increase the risk of developing chronic diseases such as heart disease, diabetes, and cancer. These conditions can contribute to a shorter lifespan.

*   Depression can lead to unhealthy behaviors such as poor diet, lack of exercise, substance abuse, and inadequate sleep, all of which are associated with increased risk of chronic diseases and mortality.

*   Depression is a major risk factor for suicide, which is a leading cause of premature death. Individuals with depression are at higher risk of suicide, especially if their depression is severe and untreated.









```python
#plotting
X = df2.Depression
plt.plot(X, y, 'o')
b, a = np.polyfit(X, y, 1)
plt.plot(X, b * X + a)
plt.xlabel('Depression')
plt.ylabel('Life Expectancy')
```

```python
y = df2['Avg_Life_Expectancy']
X = df2.Schizophrenia
```

```python
X = sm.add_constant(X)
X.head(2)
```

```python
model3 = sm.OLS(endog= y, exog = X).fit()
print(model3.summary())
```


*   A positive coefficient indicates a positive relationship. As schizophrenia increases, life expectancy increases. Schizophrenia typically occurs for individuals in their late teens to late 20s. It is very uncommon for individuals older than 40 to be diagnosed. Therefore, we can rule out that increased life expectancy means individuals at an older age are developing schizophrenia.

*   The countries with increased life expectancy, also come with an increased chance of developing schizophrenia. Factors such as stress, trauma, substance abuse, and social isolation can contribute to the development of schizophrenia. Changes in the environment, lifestyle, and social structures associated with increased life expectancy may increase the likelihood of these risk factors.

*   With increased life expectancy, more individuals with genetic predispositions have the opportunity to develop the disorder.

*   The p value is 0 indicating that there is statistical significance between the relationship.

*   Since the R-squared is 0.50, 50% of the variance in life expectancy can be explained by schizophrenia.

```python
#plotting
X = df2.Schizophrenia
plt.plot(X, y, 'o')
b, a = np.polyfit(X, y, 1)
plt.plot(X, b * X + a)
plt.xlabel('Schizophrenia')
plt.ylabel('Life Expectancy')
```
